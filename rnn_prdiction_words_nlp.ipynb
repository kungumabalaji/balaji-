```python
# Open the file in text mode ('rt') to handle lines as strings
corpus = open("alice_in_wonderland.txt", "rt").read()
documents = []  # Renamed to avoid variable shadowing

for line in corpus.splitlines():  # Iterate through lines using splitlines()
    line = line.strip().lower()  # Clean the line

    if len(line) == 0:
        continue

    documents.append(line)  # Add the processed line to the list

document = " ".join(documents)  # Combine lines into a single document
```

```python
chars = set([c for c in documents])
char2index = dict((c , i ) for i, c in enumerate(chars))
index2char = dict((i,c) for i , c in enumerate(chars))
```

```python
# So, each time, we will pick 10 characters and use the next character to predict.
SEQ_LEN = 10   #  means we will create input sequences of 10 characters at a time.
Step = 1 #means we will move by one character after creating each input sequence.

# to store the input sequences
input_chars = []
# to store the output thats the prediction of words
label_chars=[]

# BASICALLLY THIS LOOP START FROM 0 THAT IS STARTING AND GOES TO 10( STEP 10) CHARCHETRERS
for i in range(0, len(documents) - SEQ_LEN,Step):
    input_chars.append(documents[i:i+SEQ_LEN]) # ADDS THE 10 CHARTERS TO INPUT CHARTS
    label_chars.append(documents[i + SEQ_LEN]) # ADDS THE NEXT CHARCHERTS TO LABEL LIST

```
